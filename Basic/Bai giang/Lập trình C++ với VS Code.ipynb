<center><b style='color:blue;'>HƯỚNG DẪN CÀI ĐẶT MÔI TRƯỜNG CHO LẬP TRÌNH C/C++</b></center>

Để cài đặt môi trường cho lập trình C/C++ thì bạn cần:
- Tải và cài đặt Visual Studio Code: <a href="https://code.visualstudio.com/download">https://code.visualstudio.com/download</a>
- Tải và cài đặt MinGW-x64: <a href="https://www.msys2.org/">https://www.msys2.org/</a>

<b style='color:green;'>1. Cài đặt Visual Studio Code (VS Code):</b>
<br>Sau khi bạn đã tải bản installer cho VS Code thì bạn tiến hành cài đặt theo các bước sau:
- Bạn chạy file cài đặt bằng cách nhấn double-click vào file cài đặt

Bạn chọn <b>I accept the agreement</b>, sau đó chọn <b>Next</b> ở của sổ <b>License Agreement</b><br>


Khi cửa sổ <b>Select Additional Task</b> hiển thị thì bạn tiếp tục chọn <b>Next</b><br>


Chọn <b>Install</b> màn hình <b>Ready to Install</b> và chờ cho việc cài đặt hoàn tất.<br>


<b style='color:green;'>2.Cài đặt MinGW-x64</b>

- Bạn tải bản installer tại khung màu đỏ như trong hình sau:

- Chạy file cài đặt bằng cách double-click trên file installer

Bạn chọn <b>Next</b> tại màn hình sau:<br>


Ở màn hình tiếp theo thì bạn chọn thư mục để cài đặt tại <b>Browse...</b>, ở đây bạn có thể chọn giá trị mặc định như hình sau và nhấn <b>Next</b><br>


Chọn <b>Next</b> ở màn hình kế tiếp và chờ quá trình cài đặt MSYS2 hoàn tất<br>


Cuối cùng, bạn chọn như hình bên dưới:<br>


Bạn hãy copy và paste lệnh sau vào như hình bên dưới: 
<b style='color:blue;'>pacman -S mingw-w64-ucrt-x86_64-gcc</b> và nhấn <b>Enter</b><br>


Tiếp tục bạn chọn <b>Y</b> khi được hỏi <i style='color:red;'>Proceed with installation?</i> và bạn cần chờ trong vài phút để tiến hành cài đặt các package cần thiết<br>


Sau khi cài đặt xong thì bạn tiến hành kiểm tra version mà bạn đã cài đặt như sau:
- Kiểm tra version của gcc dùng lệnh sau:
<b style='color:blue;'>gcc --version</b>
- Kiểm tra version của g++ dùng lệnh sau:
<b style='color:blue;'>g++ --version</b><br>


Tiếp theo, bạn cần kiểm tra version của gcc hoặc g++ trên Command Prompt của Windows như sau:
- Nhấn tổ hợp phím <b>Windows + R</b> để mở hộp thoại Run<br>


- Nhập lệnh gcc --version hoặc g++ --version để kiểm tra version, nếu bạn nhận được kết quả như ảnh bên dưới thì đừng lo<br>


Vì bạn chỉ mới cài đặt MinGW-x64 mà chưa thêm nó vào biến môi trường nên Windows sẽ không biết được lệnh gcc hay g++ là gì cả. Và bây giờ bạn cần thêm chúng vào biến môi trường như sau: 

- Chuột phải trên <b>This PC</b> hoặc <b>My computer</b> và chọn <b>Properties</b><br>


Sau khi cửa sổ <b>Settings</b> được hiển thị thì bạn chọn <b>Advanced system settings</b><br>


Ở cửa sổ <b>System Properties</b> bạn hãy chọn <b>Environment Variables...</b><br>


Trong cửa sổ <b>Environment Variables</b>, bạn hãy chọn <b>Path</b> và sau đó chọn <b>Edit</b><br>


Bạn có nhớ thư mục cài đặt MSYS2 ở các bước trên không? Nếu nhớ thì bạn hãy copy đường dẫn đến thư mục cài đặt MSYS2 như sau:<br>Ví dụ mình đã cài đặt MSYS2 tại đường dẫn là <b>C:\msys64</b> thì mình sẽ copy đường dẫn này và quay lại của sổ <b>Edit environment variable</b>, sau đó bạn chọn <b>New</b> và paste đường dẫn bạn vừa copy vào và thêm vào sau đường dẫn giá trị <b>\ucrt64\bin</b> và nhấn <b>OK</b><br>


Sau đó bạn chọn <b>OK</b> ở các màn hình <b>Environment variables</b> và <b>System properties</b> 

- Kế tiếp bạn quay lại cửa sổ Command Prompt của Windows và tiến hành kiểm tra version của gcc và g++ như sau:<br>


- Bước tiếp theo, bạn hãy mở VS Code và nhấn tổ hợp phím <b>Ctrl + Shift + I</b> để mở Extension trên VS Code và nhập vào <b>C/C++</b>, sau đó chọn Install.<br>


- Nhấn tổ hợp phím <b>Ctrl + Shift + P</b> và nhập vào <b>C/C++: Edit configurations (UI)</b> để mở tab setting sau<br>


Và cuối cùng bạn hãy chọn setting như hình bên dưới để hoàn tất cài đặt môi trường cho lập trình C/C++:<br>
